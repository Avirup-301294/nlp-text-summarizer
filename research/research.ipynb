```python
def multiplication(a: int, b: int) -> int:
    return a*b
```

```python
multiplication(2, 3)    
```

```python
multiplication(2, "3")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def multiplication(a: int, b: int) -> int:
    return a*b

print(multiplication(2, 3))
print(multiplication(2, "3"))
```
